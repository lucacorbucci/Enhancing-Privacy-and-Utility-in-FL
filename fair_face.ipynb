# Fair Face Dataset

```python
import pandas as pd
from IPython.display import Image

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./data/fair_face/fairface_label_train.csv')
```

```python
df.head()
```

```python
grouped_data = df.groupby(['race'])
```

```python
grouped_data.size()
```

```python
for combination, group in grouped_data:
    print(combination)
    images = group['file'][0:3]
    for image in images:
        pil_img = Image(filename=f"./data/fair_face/{image}") 
        display(pil_img)
```
