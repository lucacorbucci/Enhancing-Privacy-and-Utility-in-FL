```python
import pandas as pd
from IPython.display import Image


pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("./data/celeba/train_original.csv")
```

```python
df.head()
```

### Sensitive columns

- Male (1 is male, -1 is female)
- Pale Skin

```python
grouped_data = df.groupby(["Male", "Pale_Skin"])
```

```python
grouped_data.size()
```

```python
for combination, group in grouped_data:
    print(combination)
    images = group["image_id"][0:3]
    for image in images:
        pil_img = Image(filename=f"./data/celeba/img_align_celeba/{image}")
        display(pil_img)
```
